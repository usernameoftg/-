–ü–æ–¥–≥–æ—Ç–æ–≤–∏–º –¥–∞–Ω–Ω—ã–µ –¥–∞—Ç–∞—Å–µ—Ç–∞ ferma_main.csv –∏ –æ–±—É—á–∏–º –Ω–∞ –Ω–∏—Ö –ø–µ—Ä–≤—É—é –º–æ–¥–µ–ª—å LinearRegression.

```python
# —Ñ–æ—Ä–º–∏—Ä—É–µ–º –≤—ã–±–æ—Ä–∫–∏ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã—Ö –∏ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö

# —Å–æ–∑–¥–∞–Ω–∏–µ –∫–æ–Ω—Å—Ç–∞–Ω—Ç—ã RANDOM_STATE
RANDOM_STATE = 42

X = df_ferma[['eke', 'protein', 'spo', 'kind', 'type_pasture', 'dad_kind', 'age']]
y = df_ferma['udoy']

# —Ä–∞–∑–±–∏–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –Ω–∞ —Ç—Ä–µ–Ω –∏ —Ç–µ—Å—Ç
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    random_state=RANDOM_STATE
)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
cat_col_names = ['kind', 'type_pasture', 'dad_kind', 'age']
num_col_names = ['eke', 'protein', 'spo']
```

#### OneHotEncoder() 

```python
# –≤—ã–±–µ—Ä–∏—Ç–µ –∫–ª–∞—Å—Å OneHotEncoder() –¥–ª—è –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è 
# –∑–∞–¥–∞–π—Ç–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—É sparse –∑–Ω–∞—á–µ–Ω–∏–µ False
encoder = OneHotEncoder(sparse=False, drop='first') 

X_train_ohe = encoder.fit_transform(X_train[cat_col_names])
X_test_ohe = encoder.transform(X_test[cat_col_names])
encoder_col_names = encoder.get_feature_names_out()

X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)
X_test_ohe = pd.DataFrame(X_test_ohe, columns=encoder_col_names)
```

#### StandardScaler()

```python
# –≤—ã–±–æ—Ä –º–µ—Ç–æ–¥–∞ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—è
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train[num_col_names])
X_test_scaled = scaler.transform(X_test[num_col_names])

X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_col_names)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=num_col_names)
```

```python
# –æ–±—ä–µ–¥–µ–Ω–∏–º
X_train = pd.concat([X_train_ohe, X_train_scaled], axis=1)
X_test = pd.concat([X_test_ohe, X_test_scaled], axis=1)
```

```python
# –æ–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ –∏ –ø–æ–ª—É—á–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –Ω–∞ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö
model_lr_1  = LinearRegression()
model_lr_1.fit(X_train, y_train)
predictions_1 = model_lr_1.predict(X_test)

# –ø—Ä–æ–≤–µ–¥–∏—Ç–µ –∞–Ω–∞–ª–∏–∑ –æ—Å—Ç–∞—Ç–∫–æ–≤
# –ø–µ—Ä–µ–¥ —ç—Ç–∏–º —Ä–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ –æ—Å—Ç–∞—Ç–∫–∏
residuals_1 = y_test - predictions_1

residuals_1
```

```python
r2_1 = r2_score(y_test, predictions_1)  

print("R2 LinearRegression =", r2_1)
```

```python

def plot_residuals(predictions, residuals):
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    # –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –æ—Å—Ç–∞—Ç–∫–æ–≤
    axes[0].hist(residuals, bins=20, color='blue', edgecolor='black', alpha=0.7)
    axes[0].set_title('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –æ—Å—Ç–∞—Ç–∫–æ–≤')
    axes[0].set_xlabel('–û—Å—Ç–∞—Ç–∫–∏')
    axes[0].set_ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
    axes[0].grid(True)
    median = np.median(residuals)
    axes[0].axvline(median, color='red', linestyle='-', linewidth=1)


    # –¢–æ—á–µ—á–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫ –æ—Å—Ç–∞—Ç–∫–æ–≤
    axes[1].scatter(predictions, residuals, s=50, c='blue', alpha=0.7, marker='o', edgecolors='black', linewidths=1)
    axes[1].set_title('–ì—Ä–∞—Ñ–∏–∫ –æ—Å—Ç–∞—Ç–∫–æ–≤')
    axes[1].set_xlabel('–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è')
    axes[1].set_ylabel('–û—Å—Ç–∞—Ç–∫–∏')
    axes[1].axhline(y=0, color='r', linestyle='-')
    axes[1].grid(True)

    plt.show()
```

```python
# –ø–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ –∏–∑–º–µ–Ω–µ–Ω–Ω—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
df_ferma.head()
```

–ü–æ–≤—Ç–æ—Ä–∏–º —à–∞–≥–∏ –∏–∑ –ø—É–Ω–∫—Ç–∞ 1 –∏ —Å–æ–∑–¥–∞–¥–∏–º –≤—Ç–æ—Ä—É—é –º–æ–¥–µ–ª—å: –ø–æ–¥–≥–æ—Ç–æ–≤–∏–º –¥–∞–Ω–Ω—ã–µ, –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å, –æ—Ü–µ–Ω–∏–º –µ—ë –∫–∞—á–µ—Å—Ç–≤–æ –∏ –Ω–∞–ø–∏—à–µ–º –≤—ã–≤–æ–¥—ã.

```python
# —Ñ–æ—Ä–º–∏—Ä—É–µ–º –≤—ã–±–æ—Ä–∫–∏ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã—Ö –∏ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö

# —Å–æ–∑–¥–∞–Ω–∏–µ –∫–æ–Ω—Å—Ç–∞–Ω—Ç—ã RANDOM_STATE
RANDOM_STATE = 42

X = df_ferma[['eke_sq', 'protein', 'spo_new', 'kind', 'type_pasture', 'dad_kind', 'age']]
y = df_ferma['udoy']

# —Ä–∞–∑–±–∏–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –Ω–∞ —Ç—Ä–µ–Ω –∏ —Ç–µ—Å—Ç
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    random_state=RANDOM_STATE
)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
cat_col_names = ['kind', 'type_pasture', 'dad_kind', 'age', 'spo_new']
num_col_names = ['eke_sq', 'protein']
```

```python
encoder = OneHotEncoder(sparse=False, drop='first') 

X_train_ohe = encoder.fit_transform(X_train[cat_col_names])
X_test_ohe = encoder.transform(X_test[cat_col_names])
encoder_col_names = encoder.get_feature_names_out()

X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)
X_test_ohe = pd.DataFrame(X_test_ohe, columns=encoder_col_names)
```

```python
# –≤—ã–±–æ—Ä –º–µ—Ç–æ–¥–∞ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—è
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train[num_col_names])
X_test_scaled = scaler.transform(X_test[num_col_names])

X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_col_names)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=num_col_names)
```

```python
# –æ–±—ä–µ–¥–µ–Ω–∏–º
X_train = pd.concat([X_train_ohe, X_train_scaled], axis=1)
X_test = pd.concat([X_test_ohe, X_test_scaled], axis=1)
```

```python
# –æ–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ –∏ –ø–æ–ª—É—á–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –Ω–∞ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö
model_lr_2  = LinearRegression()
model_lr_2.fit(X_train, y_train)
predictions_2 = model_lr_2.predict(X_test)

# –ø—Ä–æ–≤–µ–¥–∏—Ç–µ –∞–Ω–∞–ª–∏–∑ –æ—Å—Ç–∞—Ç–∫–æ–≤
# –ø–µ—Ä–µ–¥ —ç—Ç–∏–º —Ä–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ –æ—Å—Ç–∞—Ç–∫–∏
residuals_2 = y_test - predictions_2

residuals_2
```

**–°—Ä–∞–≤–Ω–∏–º –∫–∞—á–µ—Å—Ç–≤–æ —Ç—Ä—ë—Ö –º–æ–¥–µ–ª–µ–π –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏, —Å–æ–ø–æ—Å—Ç–∞–≤–∏–º –∏—Ö R¬≤ –∏ –æ—Å—Ç–∞—Ç–∫–∏**

**–û—Ü–µ–Ω–∏–º –∫–∞—á–µ—Å—Ç–≤–æ –∫–∞–∂–¥–æ–π –º–æ–¥–µ–ª–∏ –ø–æ —Ç—Ä—ë–º –º–µ—Ç—Ä–∏–∫–∞–º: MSE, MAE, RMSE**

–ß—Ç–æ–±—ã –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å, –∫–∞–∫–∞—è –º–æ–¥–µ–ª—å –ª—É—á—à–µ, –º–æ–∂–Ω–æ –ø—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ –æ—Ü–µ–Ω–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ –º–æ–¥–µ–ª–∏ (MSE, MAE, RMSE) –∏ —Å–¥–µ–ª–∞—Ç—å —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ –º–µ–∂–¥—É –Ω–∏–º–∏.

```python
def metrics(y_true, y_pred, num):
    
    print(f'\033[1m–ü–æ–∫–∞–∑–∞—Ç–µ–ª–∏ –º–æ–¥–µ–ª–∏ {num}:\033[0m \n')

    # MSE. C—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–∞—è –æ—à–∏–±–∫–∞ - Mean Squared Error 
    mse = mean_squared_error(y_true, y_pred) 
    print(f"MSE LinearRegression = {mse:.0f}")
    
    # MAE. –°—Ä–µ–¥–Ω—è—è –∞–±—Å–æ–ª—é—Ç–Ω–∞—è –æ—à–∏–±–∫–∞ - Mean Absolute Error
    mae = mean_absolute_error(y_true, y_pred) 
    print(f"MAE LinearRegression = {mae:.0f}")
    
    # RMSE. –ö–æ—Ä–µ–Ω—å —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–π –æ—à–∏–±–∫–∏ - Root Mean Squared Error
    rmse = mean_squared_error(y_test, y_pred, squared=False) 
    print(f"RMSE LinearRegression = {rmse:.0f}")
    
    r2 = r2_score(y_test, y_pred) 
    print(f"–ó–Ω–∞—á–µ–Ω–∏–µ R¬≤= {r2:.4f}")
```

[–ü–æ—á–∏—Ç–∞—Ç—å](https://pythonru.com/baza-znanij/sklearn-train-test-split) –º–æ–∂–Ω–æ —Ç—É—Ç

     

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
cat_col_names = ['kind', 'type_pasture', 'dad_kind', 'age', 'dad_name']
num_col_names = ['udoy', 'eke', 'protein', 'spo', 'fat_%', 'protein_%']
```

```python
encoder_vkus = OneHotEncoder(sparse=False, drop='first') 

X_train_ohe = encoder_vkus.fit_transform(X_train[cat_col_names])
X_test_ohe = encoder_vkus.transform(X_test[cat_col_names])

encoder_col_names = encoder_vkus.get_feature_names_out()

X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)
X_test_ohe = pd.DataFrame(X_test_ohe, columns=encoder_col_names)
```

```python
# –≤—ã–±–æ—Ä –º–µ—Ç–æ–¥–∞ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—è
scaler_vkus = StandardScaler()
X_train_scaled = scaler_vkus.fit_transform(X_train[num_col_names])
X_test_scaled = scaler_vkus.transform(X_test[num_col_names])

X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_col_names)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=num_col_names)
```

```python
# –æ–±—ä–µ–¥–µ–Ω–∏–º
X_train = pd.concat([X_train_ohe, X_train_scaled], axis=1)
X_test = pd.concat([X_test_ohe, X_test_scaled], axis=1)
```

```python
clf = LogisticRegression()
clf = clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

y_pred
```

```python
# –ø—Ä–æ–≤–µ—Ä–∏–º –∫–∞—á–µ—Å—Ç–≤–æ –º–æ–¥–µ–ª–∏ –ø–æ –º–µ—Ç—Ä–∏–∫–µ accuracy

acc = accuracy_score(y_test, y_pred)
print(f"–¢–æ—á–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç {acc:.2%}. –≠—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –º–æ–¥–µ–ª—å –ø—Ä–∞–≤–∏–ª—å–Ω–æ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–ª–∞ {(acc * len(y_test)):.0f} –∏–∑ {len(y_test)} –Ω–∞–±–ª—é–¥–µ–Ω–∏–π.")
```

–ü—Ä–∏ –ø–æ–∫—É–ø–∫–µ –∫–æ—Ä–æ–≤ –¥–ª—è –∑–∞–∫–∞–∑—á–∏–∫–∞, –∫–æ—Ç–æ—Ä—ã–π –∏–Ω—Ç–µ—Ä–µ—Å—É–µ—Ç—Å—è –≤–∫—É—Å–Ω—ã–º –º–æ–ª–æ–∫–æ–º, –±–æ–ª–µ–µ –∫—Ä–∏—Ç–∏—á–Ω–æ–π —è–≤–ª—è–µ—Ç—Å—è –æ—à–∏–±–∫–∞ –≤—Ç–æ—Ä–æ–≥–æ —Ä–æ–¥–∞.

–≠—Ç–æ —Å–ª—É—á–∞–π, –∫–æ–≥–¥–∞ –º–æ–¥–µ–ª—å –Ω–µ–ø—Ä–∞–≤–∏–ª—å–Ω–æ –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ—Ç, —á—Ç–æ –∫–æ—Ä–æ–≤–∞ –¥–∞—Å—Ç –≤–∫—É—Å–Ω–æ–µ –º–æ–ª–æ–∫–æ, –∫–æ–≥–¥–∞ –Ω–∞ —Å–∞–º–æ–º –¥–µ–ª–µ –º–æ–ª–æ–∫–æ –±—É–¥–µ—Ç –Ω–µ –≤–∫—É—Å–Ω—ã–º.



```python
y_proba = clf.predict_proba(X_test)[:,1]

data = pd.DataFrame(zip(y_test, y_proba),
             columns = ['y_valid', 'y_proba']).sort_values(by='y_proba',ascending=False)

data.head(5)
```

```python
from sklearn.metrics import precision_recall_curve

precisions, recalls, thresholds = precision_recall_curve(data['y_valid'], data['y_proba'])

for threshold in thresholds:
    idx = np.argmax(thresholds >= threshold)
    precision = precisions[idx]
    recall = recalls[idx]
    print(f"–ü—Ä–∏ –ø–æ—Ä–æ–≥–µ {threshold:.3f}: Precision = {precision:.3f}, Recall = {recall:.3f}")
```
