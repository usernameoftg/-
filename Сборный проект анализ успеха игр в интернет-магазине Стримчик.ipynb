Лидерами среди платформ являются PS2 и Nintendo DS, со значительным отставанием следуют PS3, Wii, X360, PSP и PS1

```python
data.pivot_table(index='platform', 
                 values='all_sales', 
                 aggfunc='sum').sort_values('all_sales', 
                                              ascending=False).plot(kind='bar', 
                                                                   figsize=(10, 5),
                                                                   title='Количество проданных игр по платформам',
                                                                   xlabel='название платформы',
                                                                   ylabel='кол-во игр');
```

Как мы видим, абсолютным лидерам продаж является платформа PS2, затем идут X360, и совсем рядом расположились PS3, Wii, DS и PS

```python
platform_sales = data.groupby('platform')['all_sales'].sum().sort_values(ascending=False)

# Выбор топ-5 платформ с наибольшими суммарными продажами
top_platforms = platform_sales.head(5).index.tolist()

# Фильтрация данных по выбранным платформам
filtered_df = data[data['platform'].isin(top_platforms)]

# Построение распределения продаж по годам для каждой платформы
for platform in top_platforms:
    platform_data = filtered_df[filtered_df['platform'] == platform]
    platform_data.groupby('year_of_release')['all_sales'].sum().plot(label=platform)

    
    
plt.title('Продажи по годам для топ-5 платформ')
plt.xlabel('Год выпуска')
plt.ylabel('Суммарные продажи')
plt.legend()
plt.grid(True)
plt.show()
```

После построения графика можно проанализировать, на протяжении какого временного периода наблюдаются пиковые значения для каждой платформы. Это позволит определить характерные сроки появления новых платформ и исчезновения старых. Обычно, жизненный цикл платформы составляет примерно 5-10 лет

```python
# Фильтрация данных за актуальный период (2011-2016 годы)
actual_period_df = data[(data['year_of_release'] >= 2014) & (data['year_of_release'] <= 2016)]

# Группировка по платформам и суммарные продажи
platform_sales = actual_period_df.groupby('platform')['all_sales'].sum().sort_values(ascending=False)

# Выбор топ-N платформ с наибольшими суммарными продажами
top_platforms = platform_sales.head(5).index.tolist()

# Фильтрация данных по выбранным платформам
filtered_df = actual_period_df[actual_period_df['platform'].isin(top_platforms)]

# Построение распределения продаж по годам для каждой платформы

for platform in top_platforms:
    platform_data = filtered_df[filtered_df['platform'] == platform]
    platform_data.groupby('year_of_release')['all_sales'].sum().plot(label=platform)
    
plt.xlim(2014, 2016)  
plt.title('Продажи по годам для топ-5 платформ (2014-2016)')
plt.xlabel('Год выпуска')
plt.ylabel('Суммарные продажи')
plt.xticks(range(2014, 2017))  # Указываем явно метки для оси x
plt.legend()
plt.grid(True)
plt.show()
```

## Шаг 5. Проверка гипотез


**Проверим следующие гипотезы**
* Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

* Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Установим порог статитстической значемости alpha 5%.



**Гипотеза 1**
* Нулевая гипотеза: средние пользовательские рейтинги платформ Xbox One и PC одинаковые
* Альтернативная гипотеза: средние пользовательские рейтинги платформ Xbox One и PC неодинаковые

```python
alpha = 0.05

results = st.ttest_ind(data_fl[data_fl['platform'] == 'XOne']['user_score'], 
                       data_fl[data_fl['platform'] == 'PC']['user_score'])

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

```python
xone_mean = data_fl[data_fl['platform'] == 'XOne']['user_score'].mean()
pc_mean = data_fl[data_fl['platform'] == 'PC']['user_score'].mean()

print('Средний рейтинг XOne:', round(xone_mean, 2))
print('Средний рейтинг PC:', round(pc_mean, 2))
```

Вывод: Значение p-value почти 84%. Таким образом, не получилось опровергнуть нулевую гипотезу. То есть, с вероятностью в 84% можно получить cходство рейтингов двух платформ

**Гипотеза 2**
* Нулевая гипотеза: средние пользовательские рейтинги жанров Action и Sports одинаковые
* Альтернативная гипотеза: средние пользовательские рейтинги жанров Action и Sports разные

```python
alpha = 0.05

results = st.ttest_ind(data_fl[data_fl['genre'] == 'Action']['user_score'], 
                       data_fl[data_fl['genre'] == 'Sports']['user_score'],
                      alternative='two-sided')

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

```python
action_mean = data_fl[data_fl['genre'] == 'Action']['user_score'].mean()
sports_mean = data_fl[data_fl['genre'] == 'Sports']['user_score'].mean()

print('Средний рейтинг Actions:', round(action_mean, 2))
print('Средний рейтинг Sports:', round(sports_mean, 2))
```

 Значение p-value = 8.2e-05 является очень низким по отношению к альфа. Поэтому мы отвергаем нулеввую гипотезу в пользу альтернативной, которая гласит, что средние пользовательские рейтинги жанров Action и Sports разные.
